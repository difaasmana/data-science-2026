```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', palette='Set2')

df = sns.load_dataset('tips')
print(df.shape)
print(df.head())
print(df.describe().round(2))
```

## Dataset
Dataset yang digunakan adalah `tips` dari library Seaborn. Dataset ini berisi informasi tagihan restoran, tip, jenis kelamin, hari, waktu makan, dan ukuran grup pelanggan.

```python
avg_by_day = df.groupby('day')['total_bill'].mean()

plt.figure(figsize=(8,5))
avg_by_day.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Rata-rata Total Bill per Hari')
plt.xlabel('Hari')
plt.ylabel('Total Bill (USD)')
plt.tight_layout()
plt.show()
```

### Grafik 1 - What? So what? Now what?
What? Grafik ini menunjukkan rata-rata total bill pada setiap hari.
So what? Terlihat bahwa ada hari tertentu dengan rata-rata tagihan lebih tinggi dibanding hari lainnya.
Now what? Restoran dapat mengevaluasi hari dengan transaksi tertinggi untuk strategi pelayanan dan promosi.

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='total_bill', hue='time', kde=True)
plt.title('Distribusi Total Bill per Waktu Makan')
plt.xlabel('Total Bill (USD)')
plt.ylabel('Frekuensi')
plt.tight_layout()
plt.show()
```

### Grafik 2 - What? So what? Now what?
What? Grafik ini menampilkan distribusi total bill berdasarkan waktu makan, yaitu lunch dan dinner.
So what? Kita bisa melihat perbedaan pola tagihan antara dua waktu makan tersebut.
Now what? Analisis ini dapat membantu melihat waktu makan mana yang cenderung menghasilkan transaksi lebih besar.

```python
plt.figure(figsize=(9,5))
sns.boxplot(data=df, x='day', y='total_bill', hue='sex')
plt.title('Distribusi Total Bill per Hari dan Gender')
plt.xlabel('Hari')
plt.ylabel('Total Bill (USD)')
plt.tight_layout()
plt.show()
```

### Grafik 3 - What? So what? Now what?
What? Boxplot ini menunjukkan sebaran total bill per hari dan dibedakan berdasarkan gender.
So what? Grafik ini membantu melihat median, rentang data, dan kemungkinan outlier pada tiap kelompok.
Now what? Restoran dapat memahami pola pengeluaran pelanggan berdasarkan kelompok yang berbeda.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', size='size', sizes=(30,180), alpha=0.7)
sns.regplot(data=df, x='total_bill', y='tip', scatter=False, color='gray')
plt.title('Hubungan Total Bill dan Tip')
plt.xlabel('Total Bill (USD)')
plt.ylabel('Tip (USD)')
plt.tight_layout()
plt.show()
```

### Grafik 4 - What? So what? Now what?
What? Scatter plot ini menunjukkan hubungan antara total bill dan tip.
So what? Terlihat adanya kecenderungan bahwa semakin besar total bill, semakin besar pula tip yang diberikan.
Now what? Insight ini dapat digunakan untuk memahami perilaku pelanggan dalam memberi tip.

## Kesimpulan
Melalui visualisasi ini, saya dapat melihat pola rata-rata tagihan, distribusi transaksi, persebaran data, dan hubungan antara total bill dengan tip. Visualisasi membantu memahami data dengan lebih cepat dan jelas dibanding hanya melihat tabel angka.
